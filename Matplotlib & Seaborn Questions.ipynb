```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
'''
Question 1 =
Subplots for Multiple Visualizations:
Use the following data to create a single figure with two subplots:
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 30, 40, 50]
y2 = [5, 15, 25, 35, 45]

Subplot 1: A line plot for y1 vs. x.
Subplot 2: A scatter plot for y2 vs. x.
Ensure both subplots have titles and labels.
'''
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 30, 40, 50]
y2 = [5, 15, 25, 35, 45]

fig = plt.subplots(1,2, figsize = (20,10))
plt.subplot(1,2,1)
sns.lineplot(x = x,y = y1, marker="o")
plt.title('Line Plot')

plt.subplot(1,2,2)
sns.scatterplot(x = x, y = y2)
plt.title('Scatter Plot')
plt.show()
```

```python
'''
Question 2 =
    Pie Chart with Percentage Labels:
    Create a pie chart to represent market share using the following data:
    labels = ['Product A', 'Product B', 'Product C', 'Product D']
    sizes = [20, 35, 25, 20]
    Display the percentages inside each slice.
    Highlight the slice with the largest share using explode.
    Add a shadow effect for better visualization.

'''
labels = ['Product A', 'Product B', 'Product C', 'Product D']
sizes = [20, 35, 25, 20]
explode = (0,0.1,0,0)
plt.pie(sizes, labels = labels, explode = explode, autopct="%0.f%%", shadow = True )
plt.title('Pie Chart')
plt.show()
```

```python
'''
Question 4 =
Analyze and Filter Data from a Pandas DataFrame:
Write a Python script to perform the following tasks on a given dataset:
Create a DataFrame:

Create a Pandas DataFrame from the following dictionary:
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 22],
    'Salary': [50000, 60000, 75000, 80000, 48000],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR']
}

Perform Data Analysis:

Calculate and display the average salary of employees.
Count the number of employees in each department.

Filter Data:

Filter and display the rows where the salary is greater than the average salary.
Filter and display rows of employees who belong to the "HR" department and are under 30 years old.
Sort Data:

Sort the DataFrame by Age in descending order and display the result.
'''
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 22],
    'Salary': [50000, 60000, 75000, 80000, 48000],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR']
}

df = pd.DataFrame(data)
df
```

```python
#Calculate and display the average salary of employees.
df['Salary'].mean()
```

```python
#Count the number of employees in each department.
department = df.groupby('Department').size()
department
```

```python
#Filter and display the rows where the salary is greater than the average salary.
df.loc[df['Salary'] > df['Salary'].mean()]
```

```python
#Filter and display rows of employees who belong to the "HR" department and are under 30 years old.

df.loc[(df['Department'] == "HR") & (df["Age"] < 30)]
```

```python
#Sort the DataFrame by Age in descending order and display the result.

data = df["Age"].sort_values(ascending=False)
data
```
